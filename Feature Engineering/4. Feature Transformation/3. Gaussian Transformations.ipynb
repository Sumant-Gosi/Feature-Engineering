# Guassian Transformations

Process of converting/transforming a given distribution into a normal(Gaussian) distribution

Some machine learning algorithms like linear and logistic assume that the features are normally distributed. 

-Accuracy
-Performance


- logarithmic transformation
- reciprocal transformation
- square root transformation
- exponential transformation (more general, you can use any exponent)
- boxcox transformation


```python
import pandas as pd
```

```python
df=pd.read_csv('titanic.csv',usecols=['Age','Fare','Survived'])
df.head()
```

```python
### fill nan
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
df.isnull().sum()
```

```python
import scipy.stats as stat
import pylab 
from matplotlib import pyplot as plt
```

```python
#### If you want to check whether feature is guassian/normal distributed
#### Q-Q plot
### (row, column, index)

def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
    
```

```python
plot_data(df,'Age')

#From the 2nd fig(Q-Q plot), it is evident that Age is normally distributed.
```

# Logarithmic Transformation

```python
import numpy as np

#Apply logarithmic transformation on the Age_log feature 
df['Age_log']=np.log(df['Age'])


plot_data(df,'Age_log')

#from 2nd fig, it can shown that by applying log transformation, we cannot a get a normal distribu
```

# Reciprocal Trnasformation

```python
#Applying reciprocal tranformation
df['Age_reciprocal']=1/df.Age

plot_data(df,'Age_reciprocal')
#from 2nd fig, it can shown that by applying reciprocal transformation, we cannot a get a normal distribu
```

# Square Root Transformation

```python
#Apply square root tranformation
df['Age_sqaure']=df.Age**(1/2)

plot_data(df,'Age_sqaure')
#from 2nd fig, it can shown that by applying square root transformation, we can a get a normal distribu
```

# Exponential Transdormation

```python
#Apply Exponential tranformation
df['Age_exponential']=df.Age**(1/1.2)

#from 2nd fig, it can shown that by applying Exponential transformation, we can a get a normal distribu
plot_data(df,'Age_exponential')
```
